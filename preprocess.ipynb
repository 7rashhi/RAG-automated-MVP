```python
import os
import glob

def extract_code_files(directory):
    """Extract all source code files from the project directory."""
    code_files = []
    for ext in ["*.py", "*.js", "*.java", "*.cpp", "*.md","*.css","*.txt"]:  
        code_files.extend(glob.glob(os.path.join(directory, "**", ext), recursive=True))
    
    return code_files


project_path = "project_repo" 
files = extract_code_files(project_path)
print(f"Extracted {len(files)} files from the project.")
print(files)  
```

```python
import ast

def extract_code_units(file_path):
    """Extract functions, classes, and docstrings from a Python file."""
    with open(file_path, "r", encoding="utf-8") as f:
        code = f.read()

    tree = ast.parse(code)
    units = []

    for node in ast.walk(tree):
        if isinstance(node, (ast.FunctionDef, ast.ClassDef)):  # Functions & Classes
            units.append(ast.unparse(node))  # Convert AST node back to code

    return units

# Example usage
test_file = "project_repo/app.py"  
units = extract_code_units(test_file)
print("\nExtracted Code Units:")
for unit in units:
    print(unit)
```
